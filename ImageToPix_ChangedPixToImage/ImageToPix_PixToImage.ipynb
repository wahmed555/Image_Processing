```python
from PIL import Image
import numpy as np 
import pandas as pd
```

Converting Image to Pixels In a CSV

```python
#creating image object
colourImg = Image.open("image.png")　

#converting image to rgb format
colourPixels = colourImg.convert("RGB") 

#converting image pixels to array and resizing array
colourArray = np.array(colourPixels.getdata()).reshape(colourImg.size + (3,))

#np.moveaxis Move axes of an array to new positions.
#Other axes remain in their original order.
#Compute an array where the subarrays contain index 
#values 0, 1, … varying only along the corresponding axis.
indicesArray = np.moveaxis(np.indices(colourImg.size), 0, 2)   
print(colourImg.size)                                                                
                                                                
                                                                
#Stack arrays in sequence depth wise (along third axis).
#-1 is for removing the axis formed when axises are moved
allArray = np.dstack((indicesArray, colourArray)).reshape((-1, 5)) 

#converting numpy array to dataframe
df_rgb_xy = pd.DataFrame(allArray, columns=["y", "x", "red","green","blue"]) 
```
